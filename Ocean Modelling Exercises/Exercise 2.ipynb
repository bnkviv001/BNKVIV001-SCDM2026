### AI Prompt

plot the analytical solution of the model  presented in eq. (4) to show the time evolution of phytoplankton and nutrient concentrations. Use the following parameters: ◦ Time axis in days, from 0 to 20 days ◦ k′ = 0.1 d−1; P0 = 0.1 mmol N m−3 and N0 = 5 mmol N m−3

k =k′N

dN/dt= − k′NP (4)

dP/dt=k′NP

*The AI generated code was slightly modified to match the equation 4 more closely

```python
# Parameters
kp = 0.1       # d^-1
P0 = 0.1       # mmol N m^-3
N0 = 5.0       # mmol N m^-3

# Time vector
t = np.linspace(0, 20, 500)

# Analytical solution
P = ((P0 + N0) * P0) / (P0 + (N0) * np.exp(-(P0+N0) * kp * t))
N = P0 + N0 - P

# Plot
plt.figure(figsize=(8,5))
plt.plot(t, P, lw=2, label='Phytoplankton (P)')
plt.plot(t, N, lw=2, label='Nutrient (N)')

plt.xlabel('Time (days)')
plt.ylabel('Concentration (mmol N m$^{-3}$)')
plt.title('Analytical Solution of the NP Model')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

### Question 1: Why are the curves idential?

The curves are identical because they are governed by the same analytical equation which states that P + N = P0 + N0 = 5.1 mmol N m^-3. This means that the total sum of P and N is conserved so if N decreases, P increases by the same amount. So it is built into the equation that the curves will have the same shape and be inverse to each other.


### Question 2: Changing the value of the uptake parameter

As shown in the plot below, increasing the value of k' increases the rate at which phytoplankton consume nutrients. This is equivalent to increasing the percentage of food consumed per day.
This results in a steeper curve and a more rapid levelling off.

```python
import numpy as np
import matplotlib.pyplot as plt

# Initial conditions
P0 = 0.1       # mmol N m^-3
N0 = 5.0       # mmol N m^-3

# Different k' values to compare
kp_values = [0.02, 0.05, 0.1, 0.2, 0.5]

# Time vector
t = np.linspace(0, 20, 500)

# Create figure
fig, axes = plt.subplots(1, 2, figsize=(12, 5), sharex=True)

# Total nitrogen
M = P0 + N0

for kp in kp_values:

    # Analytical solutions
    P = (M * P0) / (P0 + N0 * np.exp(-M * kp * t))
    N = M - P

    axes[0].plot(t, P, lw=2, label=f"k' = {kp}")
    axes[1].plot(t, N, lw=2, label=f"k' = {kp}")

# Phytoplankton panel
axes[0].set_title("Phytoplankton")
axes[0].set_xlabel("Time (days)")
axes[0].set_ylabel("P (mmol N m$^{-3}$)")
axes[0].grid(True)
axes[0].legend()

# Nutrient panel
axes[1].set_title("Nutrient")
axes[1].set_xlabel("Time (days)")
axes[1].set_ylabel("N (mmol N m$^{-3}$)")
axes[1].grid(True)
axes[1].legend()

plt.suptitle("Effect of $k'$ on Nutrient-Limited Growth")
plt.tight_layout()
plt.show()
```
